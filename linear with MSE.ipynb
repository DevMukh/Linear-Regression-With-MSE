```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
plt.rcParams['figure.figsize'] = (12.0, 9.0)

data = pd.read_csv('data.csv')
data.head
```

```python
data.info
```

```python
data.nunique
```

```python
X = data.iloc[:, 0]
Y = data.iloc[:, 1]

# Feature Scaling
X = (X - np.mean(X)) / np.std(X)

plt.scatter(X, Y)
plt.show()
```

```python
m = 5
c = 5
alpha = 0.0001  # Adjust the learning rate
epochs = 1000

count = float(len(X))
print(count)
```

```python
for i in range(epochs):
    Y_pred = m * X + c
    D_m = (-2 / count) * sum(X * (Y - Y_pred))
    D_c = (-2 / count) * sum(Y - Y_pred)
    m = m - alpha * D_m
    c = c - alpha * D_c
    print(m, c, sum(np.square(Y - Y_pred) / count), mean_squared_error(Y, Y_pred))
```

```python
Y_pred = m * X + c
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```
